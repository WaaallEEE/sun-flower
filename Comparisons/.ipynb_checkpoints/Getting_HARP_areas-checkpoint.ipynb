# Example on getting Active Region areas from SDO/HMI metadata

- more info at http://jsoc.stanford.edu/jsocwiki/HARPDataSeries. 

- Requires installing the drms python package:
https://pypi.org/project/drms/

```python
import drms
c = drms.Client()
```

### Query the metadata into a Pandas dataframe
- Description of metadata at http://jsoc.stanford.edu/jsocwiki/HARPDataSeries
- This is where you put the actual metadata that you want (e.g. AREA)

```python
k = c.query('hmi.Mharp_720s[][2014.11.30_00:00_TAI/1d@6h]{continuum}',
            key='T_REC, HARPNUM, AREA')
print(k)
```
