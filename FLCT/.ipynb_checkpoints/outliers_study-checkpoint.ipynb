```python
import os, glob
from pathlib import PurePath
import fitsio
import numpy as np
from scipy.ndimage.filters import gaussian_filter 
from scipy.signal import convolve2d
import matplotlib.pyplot as plt

def load_vel_mean(v_files, trange):
    "Load the velocity files and average over a time range"
    vxs = []
    vys = []
    vx_files_subset = v_files[0][trange[0]:trange[1]]
    vy_files_subset = v_files[1][trange[0]:trange[1]]
    for vxf, vyf in zip(vx_files_subset, vy_files_subset):
        vxs.append(fitsio.read(vxf))
        vys.append(fitsio.read(vyf))
    # Get the mean of the velocity components
    vx = np.array(vxs).mean(axis=0)
    vy = np.array(vys).mean(axis=0)
    return vx, vy


def smooth_vel(vx, vy, fwhm, kernel='boxcar'):
    """ Smooth the velocity with a smoothing kernel that can either be: 
     - boxcar: width set to fwhm
     - gaussian: parametrized by fwhm. 
     
     Returns the smoothed velocity components
    """ 
    
    if kernel == 'boxcar':
        box = np.ones([fwhm, fwhm]) / fwhm**2
        vxs2 = convolve2d(vx, box, mode='same')
        vys2 = convolve2d(vy, box, mode='same')
    elif kernel == 'gaussian':
        sigma = fwhm / 2.35
        vxs2 = gaussian_filter(vx, sigma=sigma, order=0)
        vys2 = gaussian_filter(vy, sigma=sigma, order=0)
        
    return vxs2, vys2
```

```python
parentdir =PurePath(os.environ['DATA'], 'Ben', 'SteinSDO', 'FLCT_Raphael')
prefix = 'output_FLCT_'
fdirs = [PurePath(parentdir, prefix+subdir) for subdir in ['sigma4', 'sigma15', 'sigma4_skip8', 'sigma15_skip8']]
ffits_x = [sorted(glob.glob(PurePath(fdir, 'FLCT_vx*.fits').as_posix())) for fdir in fdirs]
ffits_y = [sorted(glob.glob(PurePath(fdir, 'FLCT_vy*.fits').as_posix())) for fdir in fdirs]
# Make list of files in this order [fdirs, (x,y), frame #]
ffits = list(zip(ffits_x, ffits_y))
trange = [0, 60]
# unit in km/s
u = 368 / 60 
```

```python
vsigmas = [load_vel_mean(fdir, trange) for fdir in ffits]
```

```python
vmagsigmas = [np.sqrt(vsigma[0]**2 + vsigma[1]**2).ravel()*u for vsigma in vsigmas]
# Smooth the 1st one from sigma = 4 to 15*1.6 px. (sigma = 15 in FLCT means FWHM = 15*1.6)
v0sm_x, v0sm_y = smooth_vel(vsigmas[0][0]*u, vsigmas[0][1]*u, round(15*1.6))
v0_smooth_mag = np.sqrt(v0sm_x**2 + v0sm_y**2).ravel()
vmagsigmas.append(v0_smooth_mag)
```

```python
vhists, binEdges=zip(*[np.histogram(v.ravel(), bins=100) for v in vmagsigmas])
bincenters = [0.5*(bins[1:]+bins[:-1]) for bins in binEdges]
```

```python
%matplotlib inline
plt.figure(figsize=(8,8))
plt.plot(bincenters[0], vhists[0], color='black', label='FLCT @sigma = 4 px')
plt.plot(bincenters[1], vhists[1], color='red', label='FLCT @sigma = 15 px')
plt.plot(bincenters[2], vhists[2], color='black', ls='--', label='FLCT @sigma = 4 px smoothed to sigma = 15 px')
plt.xlabel('velocity magnitude [km/s]')
plt.legend()
```
